```python
dummies=pd.get_dummies(df[['TotalSteps']],drop_first=True)
data_to_be_scaled=df.drop(['TotalSteps'],axis=1)

df_feat=data_to_be_scaled
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df_feat)

df_scaled=scaler.transform(df_feat)
df_scaled=pd.DataFrame(df_scaled,columns=df_feat.columns[2])
df=pd.concat([df_scaled,dummies],axis=1)
df.head()
```
