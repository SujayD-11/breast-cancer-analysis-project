```python
dummies=pd.get_dummies(df[['TotalSteps','ActivityDay']],drop_first=True)
data_to_be_scaled=df.drop(['TotalSteps','ActivityDay'],axis=1)

df_feat=data_to_be_scaled
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df_feat)

df_scaled=scaler.transform(df_feat)
df_scaled=pd.DataFrame(df_scaled,columns=df_feat.columns[0:])
df=pd.concat([df_scaled,dummies],axis=1)
df.head()
```

```python
df.to_csv(r'../Data/preprocessedData.csv')
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(df,test_size=0.2)
```

```python
train.to_csv(r'../Input/train.csv')
test.to_csv(r'../Input/test.csv')
```
